```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples = 500, n_features = 2, n_redundant = 0, n_classes = 2,class_sep = 1.25, random_state = 77)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap = 'winter',s=50)
```

## Train Test and Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred_logistic = clf.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred_logistic)
print("accuracy of the logistic regression is:" , score*100,"%")
```

```python
print("Classification report of logistic regression:")
print(classification_report(y_test,y_pred_logistic))
```

```python
print("Confusion matrix of the given regression is: ")
print(pd.DataFrame(confusion_matrix(y_test,y_pred_logistic)))
```

```python
m = clf.coef_
b = clf.intercept_
print("The coefficients are:",m)
print("The intercept is:", b[0])
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred_linear = model.predict(X_test)
```

```python
score = r2_score(y_test,y_pred_linear)
print("Accuracy of the linear regression is:",score*100)
```
